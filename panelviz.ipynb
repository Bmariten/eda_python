```python
# import libraries
import matplotlib.pyplot as plt 
import pandas as pd 
import numpy as np
```

```python
# Getting the data 
df= pd.read_csv('data/ToyotaCorolla.csv')

# plotting a scatter to show the relationship between Age and Price

fig, ax = plt.subplots()
plt.scatter(df['Age'], df['Price'])
ax.set_title('Age vs Price of Toyota cars')
ax.set_xlabel('Age Cars')
ax.set_ylabel('Price of the Cars')
plt.show()
```

```python
# Plotting a histogram - 
%matplotlib inline
plt.hist(df['Price'],density=True, histtype='bar', facecolor='g')
plt.title('Histogram Example')
plt.xlabel('Data')
plt.ylabel('Frequency')
plt.show()
```

```python
# Creating Bar Graph - used to compare & contrast of different data points ( Categorical Data)
# loading the players data 
players=pd.read_csv('data/players.csv')
plt.bar("Player", "Stat", data=players, color='green')
plt.xlabel('Players')
plt.ylabel('Goals Scored')
plt.title('Goal Scorers in EPL')
```

SEABORN LIBRARY

```python
# Scatter Pot - We use it to show LINEAR relationship between data points
#import Seaborn library
import seaborn as sns

plt.style.use("ggplot")
plt.figure(figsize=(8,8))
sns.regplot(x=df["Age"], y=df["Price"])
#replot is regression plot 
plt.show()
```

```python
# Adding a third variable using 'hue'

sns.Implot(x="Age", y="Price", data=df, fit_reg=False, hue="FuelType", legend=True, palette='Set1', height=8)
```

```python
# Plotting a histogram
plt.figure(figsize=(8,8))
sns.displot(df['Age'], kde=True, bins=5)
plt.ylabel('Frequency')
plt.show()
```
